    You are using the tf.keras.layers.experimental.preprocessing.Rescaling layer, which has not yet been ported to TensorFlow JS.

```python
# %load batch_classification.py
import os
import shutil
from keras.preprocessing import image
import numpy as np
from keras.applications.imagenet_utils import preprocess_input
from tqdm import tqdm

# move images predicted as y = 0 from src to dest
def batch_binary_classification(model, src_dir, target_dir, labels, img_dim = 224):
    
    for l in labels:
        os.makedirs(os.path.join(target_dir, l), exist_ok=True) # create the target sub folders if not exist

    
    for root, dirs, files in tqdm(os.walk(src_dir)):      
        for file in files:
            if file.endswith('.jpg') or file.endswith('.JPG') or file.endswith('.png'):
                img_path = os.path.join(root, file)
                try: 
                    img = image.load_img(img_path, target_size=(img_dim, img_dim))
                    x = image.img_to_array(img)
                    x = np.expand_dims(x, axis=0)
                    x = preprocess_input(x)
                    p = model.predict(x)                
                    cid = int(p.argmax(-1))
                    cname = labels[cid]

                    if(cid==0): # and p[0][0] > threshold):
                        shutil.move(img_path, os.path.join(target_dir, cname, file))
                except Exception as err: 
                    print(err)                
```
